```python
# import pygame
import pygame

# initialize game engine
pygame.init()

window_width=700
window_height=700

animation_increment=10
clock_tick_rate=20

# Open a window
size = (900, 700)
screen = pygame.display.set_mode(size)

# Set title to the window
pygame.display.set_caption("Covid USA")

dead=False

clock = pygame.time.Clock()
background_image = pygame.image.load("NC.jpg").convert()

while(dead==False):
    background_image= pygame.transform.smoothscale(background_image, (900,700))
    
    for event in pygame.event.get():

        if event.type == pygame.MOUSEBUTTONUP:
            start = event.pos
            print("start",start)

        if event.type == pygame.QUIT:
            dead = True
    screen.blit(background_image, [0, 0])
    pygame.draw.circle(screen,(155,0,0),(890,250),(1))
    pygame.display.flip()
    pygame.display.update()
    clock.tick(clock_tick_rate)
```
